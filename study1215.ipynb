### 예제1 : age의 결측치를 age의 중앙값(median) 으로 채워서 age_filled 컬럼을 만드세요.

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "name": ["A", "B", "C", "D", "E", "F"],
    "age":  [23,  np.nan, 31, 28,  np.nan, 40],
    "score":[80,  90,    np.nan, 70,  85,   np.nan],
    "city": ["Seoul", None, "Busan", None, "Seoul", "Incheon"]
})
```

```python
print(df)
```

```python
df_median = df['age'].median()
newdf = df['age'].fillna(df_median)

df['age_filled'] = newdf

print(df)
```

### 예제2 : score의 결측치를 score의 평균(mean) 으로 채워서 score_filled 컬럼을 만드세요.

```python
score_mean = df['score'].mean()


df['score_filled'] = df['score'].fillna(score_mean)

print(df)
```

### 예제3 : city의 결측치를 문자열 "Unknown"으로 채워서 city_filled 컬럼을 만드세요.

```python
df['city_filled'] = df['city'].fillna('Unknown')
print(df)
```
